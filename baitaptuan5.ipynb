```python
# bai 1
def Max_Of_Two( a, b) :
    if a > b:
        return a
    elif a==b:
        return a
    else:
        return b

a,b = map(int,input("nhap vao hai so nguyen a va b: ").split())
print(Max_Of_Two(a,b))
```

```python
# bai 2
def swap(a,b):
    a , b = b, a
    return a ,b

a,b = map(int,input("nhap hai so nguyen a va b: ").split())
print(*swap(a,b))
```

```python
# bai 3
import math
def songuyento(n):
    if n == 1 and n == 2:
        return False
    else:
        for i in range(2, int(math.sqrt(n))+1):
            if n % i == 0:
               return False
        return True

n=int(input("nhap vao so nguyen: "))
print(songuyento(n))
```

```python
# bai 4
def sohoanhao(n):
    s=0
    for i in range(1,n+1):
        if n % i == 0:
            s += i
        if s == n:
            return True
    return False

print(sohoanhao(28))
```

```python
# bai 5
def timso(danhsach,a):
    for i in range(len(danhsach)):
        if danhsach[i] == a:
            return danhsach[i]
    return -1

danhsach=[1,2,3,4,-6,12,45,67,89]
a = 5
print(timso(danhsach,a))
```

```python
# bai 6
def giaithua(n):
    t= 1
    for i in range(1,n+1):
        t *=i
    return t


print(giaithua(14))
```

```python
# bai 7
def maytinhbotui(a,operat,b):
    if operat == "+":
        return f"{a + b:.2f}"
    elif operat == "-":
        return f"{a - b:.2f}"
    elif operat == "*":
        return f"{a * b:.2f}"
    elif operat == "/":
        return f"{a / b:.2f}"

print(maytinhbotui(23.89,"*",23.12))
```

```python
# bai 8
def Hamming(a,b):
    c= bin( a ^ b)
    s= 0
    for i in range(0,len(c)):
        if c[i] == "1":
            s += 1
        else:
            s += 0
    return s

print(Hamming(123456,32321))
```

```python
# bai 9
def tongchuso(n):
    n=str(n)
    s=0
    for i in range(0,len(n)):
        s = s + int(n[i])
    return s

n=int(input("nhap vao so nguyen duong: "))
print(tongchuso(n))
```

```python
# bai 10
def anhxa(a,b)-> bool:
    map_a_to_b = {}
    map_b_to_a = {}
    for phantu_a,phantu_b in zip(a,b):
        if phantu_a in map_a_to_b :
            if map_a_to_b[phantu_a] != phantu_b:
                return False
        elif phantu_b in map_b_to_a :
            if map_b_to_a[phantu_b] != phantu_a:
                return False
        else:
            map_a_to_b[phantu_a] = phantu_b
            map_b_to_a[phantu_b] = phantu_a
    return True

print(anhxa("egg","add"))
```

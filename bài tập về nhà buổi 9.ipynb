```python
# bai 1
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]

        merge_sort(left_half)
        merge_sort(right_half)

        i = j = k = 0
        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
               arr[k] = left_half[i]
               i += 1
            else:
               arr[k] = right_half[j]
               j += 1
            k += 1
        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1
        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1
    return arr

arr = list(map(int,input().split()))
print(merge_sort(arr))
```

```python
# bai 2
def merge(a , b , i ,j , ans):
    if (i == len(a)):
        for k in range(j , len(b)):
            ans.append(b[k])
        return ans
    elif (j == len(b)):
        for k in range(i , len(a)):
            ans.append(a[k])
        return ans
    else:
        if (a[i] < b[j]):
            ans.append(a[i])
            return merge(a , b , i + 1 , j , ans)
        else:
            ans.append(b[j])
            return merge(a , b , i , j + 1 , ans)
k = 0
a = []
while True:
    line = list(map(int , input().split()))
    if not line:
        k = a[len(a) - 1][0]
        a.pop()
        break
    else:
        a.append(line)
ans = []
for row in a:
    ans = merge(ans , row , 0 , 0 ,[])
print(ans[k-1])
```

```python
# bai 3
def tim_so_nho_nhat(n_str):
    s = list(str(n_str))
    length = len(s)
    
    for i in range(length - 1):
        min_val = s[i]
        idx_swap = -1
        
        for j in range(i + 1, length):
            if i == 0 and s[j] == '0':
                continue
            
            if s[j] < s[i] and s[j] <= min_val:
                min_val = s[j]
                idx_swap = j
        
        if idx_swap != -1:
            s[i], s[idx_swap] = s[idx_swap], s[i]
            return "".join(s)
    
    seen = set()
    for char in s:
        if char in seen:
            return "".join(s) 
        seen.add(char)
        
    s[length-1], s[length-2] = s[length-2], s[length-1]
    
    return "".join(s)

n = int(input())
print(tim_so_nho_nhat(n))
```

```python
# bai 7
def canPlaceFlowers(place ,i ,k):
    if (k == 0):
        return True
    if (i == len(place)):
        return False
    if (place[i] == 0):
        checkleft = True
        checkright = True
        if (i != 0) and (place[i - 1] != 0):
            checkleft = False
        if (i != len(place) - 1) and (place[i + 1] != 0):
            checkright = False
        if (checkleft and checkright):
            place[i] = 1
            return canPlaceFlowers(place , i + 1 , k - 1)
    return canPlaceFlowers(place , i + 1 , k)
place = list(map(int , input().split()))
k = int(input())
print(canPlaceFlowers(place , 0 , k))
```

```python
# bai 8
map = {"R" : [1 , 0] , "L" : [-1 , 0] , "U" : [0 , 1] , "D" : [0 , - 1]}
def is_path_crossing(i , moves , x , y , crossed):
    if (crossed[x][y]):
        return True
    if (i == len(moves)):
        return False
    crossed[x][y] = True
    u , v = map[moves[i]]
    return is_path_crossing(i + 1, moves , x + u , y + v , crossed)
moves = input()
crossed = [[False for _ in range(len(moves)*2 + 1)] for _ in range(len(moves)*2 + 1)]
print(is_path_crossing(0 , moves , len(moves) , len(moves) , crossed))
```

```python
# bai 10
def findRadius(i , j ,houses , heaters):
    if (i == len(houses)):
        return 0
    if (houses[i] < heaters[j]):
        d = abs(houses[i] - heaters[j])
        if (j != 0):
            d = min(d , houses[i] - heaters[j - 1])
        return max(d , findRadius(i + 1 , j , houses , heaters))
    if (j == len(heaters) - 1):
        return max(abs(houses[i] - heaters[j]) , findRadius(i + 1 , j , houses , heaters))
    return findRadius(i , j + 1 , houses , heaters)
houses = list(map(int , input().split()))
heaters = list(map(int , input().split()))
houses.sort()
heaters.sort()
print(findRadius(0 , 0 ,houses , heaters))
```

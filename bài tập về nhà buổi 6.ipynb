```python
# bai 5
n=map(int,input("nhap cac so nguyen: ").split())
d={
   'positives' : 0,
   'negative' : 0,
   'zero' : 0
}

for x in n:
    if x > 0:
       d['positives'] +=  1
    elif x < 0:
       d['negative'] += 1
    else:
       d['zero'] += 1

print(d)

        
```

```python
# bai 9
d1 = eval(input())
d2 = eval(input())

d = {}

for key in d1:
    d[key] = d1[key]

for key in d2:
    if key in d1:
        d[key] += d2[key]
    else:
        d[key] = d2[key]

ds = dict(sorted(d.items()))

print(ds)
```

```python
# bai 12
n = list(map(int,input("nhap cac so nguyen: ").split()))
d = {}
for x in n:
    key = x
    if key in d:
        d[key] += 1 
    else:
        d[key] = 1 

b=list(d.items())
b.sort(key = lambda x : (-x[1], x[0]))
print(b[0][0])
```

```python
# bai 16
m,n = map(int,input().split())
matrix=[]
for i in range(n):
    row = list(map(int,input().split()))
    if len(row) != m:
        print("nhap lai m,n")
    matrix.append(row)
for row in matrix:
    for v in row :
        print(f"{v:>4}",end=" ")
    print()

```

```python
# bai 18
m,n,k=map(int,input("nhap so nguyen: ").split())
matrix=[]
for i in range(n):
    row = list(map(int,input().split()))
    if len(row) != m:
        print("nhap lai")
    matrix.append(row)
tong1=0
tong2=0
for i in range(n):
    tong1 += matrix[k][i]
print(tong1)

for j in range(m):
    tong2 += matrix[j][k]
print(tong2)
```

```python
# bai 20
n=list(map(int,input("nhap cac so nguyen: ").split()))
k=int(input())
def timso(n,k):
    for i in range(len(n)):
        if n[i] == k:
            return i+1
    return -1

print(timso(n,k))
        
```

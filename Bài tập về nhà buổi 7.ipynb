```python
# bai 1
def binary_search(a,l,r,x):
    while l <= r:
        m = (l+r)//2
        if a[m] == x:
            return m
        elif a[m] < x:
            l = m + 1
        else:
            r = m - 1
    return -1

a=list(map(int,input().split()))
k=int(input())
print(binary_search(a,0,len(a)-1,k))
```

```python
# bai 2
def count_occurrences(arr, x) -> int:
    dem = 0
    for i in arr:
        if i == x:
            dem += 1
    return dem 

arr=list(map(int,input().split()))
x=int(input())
print(count_occurrences(arr,x))
```

```python
# bai 3
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1,n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0,n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr

arr=list(map(int,input().split()))
print(*selection_sort(arr))
print(*bubble_sort(arr))
```

```python
# bai 8
a = eval(input())
q = eval(input())
limit = 0
LargestVal = 10**6 + 7
ans = []
for i in a:
    limit = max(limit,i[0] , i[1])
for i in q:
    limit = max(limit , i)
st =[LargestVal] * (limit << 2 | 1)
def build (i , l , r , u , v , x):
    if (u<= l) and (r<=v):
        #print(i ,l , r , u , v , x)
        st[i] = min(st[i] , x)
    elif not ((r< u) or (v < l)) : 
        mid = (l + r) >>1
        build(i << 1 , l , mid,u , v ,x)
        build(i << 1 | 1 ,mid + 1 , r , u , v ,x)
def spread(i , l , r):
    if (l != r):
        #print(l , r , st[i>>1] , st[i >> 1 | 1] , st[i])
        st[i << 1] = min(st[i << 1] , st[i])
        st[i << 1 | 1] = min(st[i << 1 | 1], st[i])
        mid = (l + r) >>1
        spread(i << 1 , l , mid)
        spread(i << 1 | 1 , mid + 1 , r)
def take(i , l , r , u):
    if ( (l != r)  and (l<= u) and (u<=r) ):
        mid = (l + r) >>1
        take(i << 1 , l , mid , u)
        take(i << 1 | 1 , mid + 1 , r , u)
    elif ( (l == r) and (l == u) ):
        if st[i] == LargestVal:
            ans.append(-1)
        else:
            ans.append(st[i])
for i in a:
    build(1 , 1 , limit , i[0] , i[1] , i[1] - i[0] + 1)
spread(1 , 1 , limit)
for i in q:
    take(1,1,limit,i)
print(ans)
```

```python
# bai 10 
import bisect

def max_subarray_mod(nums, m):
    max_mod = 0
    curr_prefix = 0
    sorted_prefixes = [0]

    for num in nums:
        curr_prefix = (curr_prefix + num) % m
        max_mod = max(max_mod, curr_prefix)
        idx = bisect.bisect_right(sorted_prefixes, curr_prefix)

        if idx < len(sorted_prefixes):
            prev_prefix = sorted_prefixes[idx]
            max_mod = max(max_mod, (curr_prefix - prev_prefix + m) % m)

        bisect.insort(sorted_prefixes, curr_prefix)

    return max_mod

nums = list(map(int,input().split()))
m = int(input())
print(max_subarray_mod(nums, m))        
```

```python
# bai 11
def tansuattu(cau,tu):
     
    d = {}
    for ch in cau:
        d[ch] = d.get(ch,0) + 1
    return d.get(tu, 0)

cau = input().split()
tu = input()

print(tansuattu(cau,tu))

```

```python
# bai 12
def sap_xep_nhanh():
    labels = ['A', 'B', 'C', 'D']
    phuong_an = []

    print("Nhập 4 phương án (mỗi phương án 1 dòng):")
    for i in range(4):
        try:
            text = input().strip()
            phuong_an.append((text, labels[i]))
        except EOFError:
            break
    
    phuong_an.sort(key=lambda x: x[0])
    ket_qua = [item[1] for item in phuong_an]

    print("Thứ tự đúng:")
    print(" ".join(ket_qua))


if __name__ == "__main__":
    sap_xep_nhanh()


```

```python
# bai 13
def longest_harmonious_subsequence(arr):
    freq = {}

    for x in arr:
        freq[x] = freq.get(x, 0) + 1

    max_len = 0
    
    for x in freq:
        if x + 1 in freq:
            max_len = max(max_len, freq[x] + freq[x + 1])

    return max_len

arr = list(map(int, input().split()))
print(longest_harmonious_subsequence(arr))
```
